```python
q1 = data['AMT_INCOME_TOTAL'].quantile(0.25)
q3 = data['AMT_INCOME_TOTAL'].quantile(0.75)
iqr = q3 - q1
low = q1 - (1.5 * iqr)
high = q3 - (1.5 * iqr)

data[(data['AMT_INCOME_TOTAL'] < low) | (data['AMT_INCOME_TOTAL'] > high)]
```

```python
print(data[(data["AMT_INCOME_TOTAL"] < low) | (data["AMT_INCOME_TOTAL"] > high)].index)
```

```python
def outlierThreshold(data, column, q1 = 0.25, q3 = 0.75):
    quantile1 = data[column].quantile(q1)
    quantile3 = data[column].quantile(q3)
    iqr = quantile3 - quantile1
    upper_limit = quantile3 + (1.5 * iqr)
    lower_limit = quantile1 - (1.5 * iqr)
    return lower_limit, upper_limit
```

```python
def checkOutlier(data, column):
    ll, ul = outlierThreshold(data = data, column = column)
    if data[(data[column] > ul) | (data[column] < ll)].any(axis = None):
        return True
    else:
        return False
```

```python
print(outlierThreshold(data = data, column = 'AMT_INCOME_TOTAL'))
```

```python
def grabColNames(dataframe, cat_th=10, car_th=20):
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O" and col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cat)}")

    return cat_cols, num_cols, cat_but_car
```

```python
cat_cols, num_cols, cat_but_car = grabColNames(data)
```

```python
for column in num_cols:
    print(column, checkOutlier(data = data, column = column))
```

```python
low, up = outlierThreshold(data = data, column = 'AMT_CREDIT')
print(data.shape)
```

```python
print(data[~((data["AMT_CREDIT"] < low) | (data["AMT_CREDIT"] > up))].shape)
```

### Replacing outliers with threshold values

```python
def replaceWithThresholds(dataframe, variable):
    low_limit, up_limit = outlierThreshold(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

```python
num_cols.remove('SK_ID_CURR')
```

```python
for col in num_cols:
    print(col, checkOutlier(data = data, column = col))
```

```python
for columns in num_cols:
    replaceWithThresholds(dataframe = data, variable = columns)
```

```python
df = sns.load_dataset('diamonds')
print(df.shape)
```

```python
df.head()
```

```python
df = df.select_dtypes(include=['float64', 'int64'])
df = df.dropna()
print(df.shape)
print(df.head())
```

```python
for col in df.columns:
    print(col, checkOutlier(df, col))
```

```python
low, up = outlierThreshold(df, "carat")
print(df[((df["carat"] < low) | (df["carat"] > up))].shape)
```

```python
low, up = outlierThreshold(df, "depth")
print(df[((df["depth"] < low) | (df["depth"] > up))].shape)
```

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=20)
clf.fit_predict(df)
df_scores = clf.negative_outlier_factor_
print(df_scores)
```

```python
scores = pd.DataFrame(np.sort(df_scores))
scores.plot(stacked=True, xlim=[0, 20], style='.-')
plt.show()
```
